```python
import requests

def fetch_news(api_key):
    # URLs for global top headlines and technology news
    url_global = f"https://newsapi.org/v2/top-headlines?country=us&apiKey={api_key}"
    url_tech = f"https://newsapi.org/v2/top-headlines?category=technology?country=us&apiKey={api_key}"
    
    
    # Fetch global top headlines
    response_global = requests.get(url_global)
    news_data_global = response_global.json()
    articles_global = news_data_global.get('articles', [])
    
    # Fetch technology news
    response_tech = requests.get(url_tech)
    news_data_tech = response_tech.json()
    articles_tech = news_data_tech.get('articles', [])
    
    # Combine articles from both calls
    articles = articles_global + articles_tech

    # Print the news articles
    for article in articles:
        title = article['title']
        description = article['description'] or 'No description available.'
        print(f"{title}\n{description}\n")

def create_app(api_key):
    fetch_news(api_key)

if __name__ == "__main__":
    api_key = 'api_key'  # Your NewsAPI key
    create_app(api_key)
```
